<font size="5"> Define the sigmoid function to use it in the Logistic Regression
    $$\sigma(z)=\frac{1}{1+\exp(-z)}$$

```python
#Define the sigmoid function
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
x=np.linspace(-10,10,1000)
plt.figure(figsize=(10,7))
plt.plot(x,sigmoid(x),'k',label='Sigmoid Function')
plt.grid()
plt.legend()
```

<font size="5"> Now let's implement the cost function and the gradient. Recall that the logistic function in logistic regression is defined as 
    $$J(\theta)=\frac{1}{m}\sum_{i=1}^{m} [-y^{i}\log(h_{\theta}(x^{i})) - (1-y^{i})\log(1-h_{\theta}(x^{i})) ]$$

```python
Mat=Data[:,0:2]
addC=np.ones((Data[:,0:2].shape[0],1))
X=np.concatenate((addC,Mat),axis=1)
y=Data[:,2]
```

```python
def cost(th,X1,y1):
    m=X1.shape[0]
    J=(1/m)*np.sum((-(y1)*np.log(sigmoid(np.dot(X1,th))) -(1-y1)*np.log(1-sigmoid(np.dot(X1,th)))  ))
    return J
```

```python
Data.shape[0]
theta=np.zeros((Data.shape[1]))
```

```python
Cs=cost(theta,X,y)
```
